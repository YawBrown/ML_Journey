```python
import pandas as pd
```

```python
raw_csv_data=pd.read_csv('Absenteeism_data.csv')
```

DATA PREPROCESSING


```python
# Make a copy of the dataset
data_set=raw_csv_data.copy()
```

```python
#Ensuring there is no maximum rows and columns and that all rows and columns in data frame are displayed
pd.options.display.max_columns=100
pd.options.display.max_rows=100
```

```python
#Giving information about the dataset.
data_set.info()
```

ABOUT TO ATTEMPT A LOGISTIC REGRESSION.

```python
#Drop ID(Label Variable,nominal data with numeric value) as the id number it doesn't affect the absenteeism time in hours
data_set.drop(['ID'],axis=1)
#axis 0 is vertical axis and axis 1 is horizontal axis
```

```python
data_set=data_set.drop(['ID'],axis=1)
```

```python
data_set
```

```python
sorted(data_set['Reason for Absence'].unique())
```

```python
reason_columns=pd.get_dummies(data_set['Reason for Absence'])
```

```python
reason_columns
```

```python
reason_columns=pd.get_dummies(data_set['Reason for Absence'], drop_first= True)
```

```python
reason_columns
```

```python
data_set.info()
```

```python
data_set=data_set.drop(['Reason for Absence'], axis = 1)
```

```python
data_set.info()
```

```python
#Group the variables
Diseases=reason_columns.loc[:, 1:14].max(axis=1)
Birth_related_issues=reason_columns.loc[:,15:17].max(axis=1)
Poisoning=reason_columns.loc[:,18:21].max(axis=1)
Light_issues=reason_columns.loc[:,22:28].max(axis=1)
```

```python
Diseases
```

```python
data_set=pd.concat([data_set,Diseases,Birth_related_issues,Poisoning,Light_issues],axis=1)
```

```python
data_set
```

```python
data_set.columns.values
```

```python
Column_names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours','Diseases' ,'Birth_related_issues','Poisoning','Light_issues']
```

```python
data_set.columns= Column_names
```

```python
data_set
```

```python
column_names_new=['Diseases' ,'Birth_related_issues','Poisoning','Light_issues','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
data_set=data_set[column_names_new]
```

```python
data_set.head()
```

```python
#checkpoint of data
df_checkpoint=data_set.copy()
```

```python
df_checkpoint['Date']=pd.to_datetime(df_checkpoint['Date'],format='%d/%m/%Y')
```

```python
type(df_checkpoint['Date'][0])
```

ADDING THE MONTH TO DATA

```python
months=[]
for i in range(700):
  months.append(df_checkpoint['Date'][i].month)
```

```python
df_checkpoint['Month_Value']=months
```

```python
len(months)
```

```python
df_checkpoint.head()
```

```python
def date_to_weekday(date_value):
  return date_value.weekday()
```

```python
df_checkpoint['Day of the week']=df_checkpoint['Date'].apply(date_to_weekday)
```

```python
df_checkpoint.head()
```

```python
df_checkpoint2=df_checkpoint.copy()
```

```python
df_checkpoint2=df_checkpoint2.drop(['Date'],axis=1)
```

```python
df_checkpoint2.columns.values
```

```python
df_checkpoint2=df_checkpoint2[['Diseases', 'Birth_related_issues', 'Poisoning', 'Light_issues','Month_Value',
       'Day of the week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']]
```

```python
df_checkpoint2.head()
```

```python
df_checkpoint3=df_checkpoint2.copy()
```

Addressing Education

```python
df_checkpoint3['Education']=df_checkpoint3['Education'].map({1:0,2:1,3:1,4:1})
```

```python
df_checkpoint3['Education'].value_counts()
```

FINAL CHECKPOINT

```python
FINAL_DATASET_FOR_ML= df_checkpoint3.copy()
```

EXPORTING TO CSV

```python
FINAL_DATASET_FOR_ML.to_csv('Absenteeism_preprocessed.csv',index=False)
```
